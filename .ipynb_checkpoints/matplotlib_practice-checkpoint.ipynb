```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

Skill Drill 5.1.7 Scatter Plot

```python
# Create the plot
plt.plot(y_axis, x_axis, 'o', c='red', label='Chicago')

plt.xlabel("Fare($)")
plt.ylabel("Date")

# Create a title.
plt.title("PyBer Fare by Month")
plt.legend()

#Add limit for the x-axis data
plt.xlim(0, 50)

#Invert the y-axis so that January is at the top
plt.gca().invert_yaxis()
```

```python
# Create a scatter plot using plt.scatter function
plt.scatter(x_axis, y_axis)
```

```python
# adjust the size of the markers
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)
```

```python
# Set new y-axis list for the parameter
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# Refactor the format for list comprehension. new_list = [expression for item in list if conditional]
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
# Create a Scatter Plot Using the Object-Oriented Interface
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

Skill Drill 5.1.7 Bubble Chart

```python
# Create a Bubble Chart Using the Object-Oriented Interface
fig, ax = plt.subplots()

plt.xlabel("Fare($)")
plt.ylabel("Date")

# Create a title.
plt.title("PyBer Fare by Month")

ax.scatter(y_axis, x_axis, s = [i * 5 for i in y_axis], label = "Boston", 
           facecolors="xkcd:sky blue", linewidth = 2, edgecolors="black", alpha=0.80)

#Invert the y-axis so that January is at the top
plt.gca().invert_yaxis()
plt.legend()
```

```python
# Creating a pie chart using same data as before
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
# Offset one of the pie slices
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
plt.show()
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", 
          "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
# Create a Pie Chart Using the Object-Oriented Interface
fig, ax = plt.subplots()
ax.pie(y_axis,labels=x_axis)
plt.show()
```

Skill Drill 5.1.8 Object-Oriented Approach

```python
#Add a percentage to one decimal place to each wedge
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", 
          "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

plt.subplots(figsize=(8, 8))

plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    shadow=True,
    startangle=-95,
    autopct='%.1f%%')

plt.gca().invert_yaxis()
plt.show()
```
